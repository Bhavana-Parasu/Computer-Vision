```python
# Create a dictionary to map the labels to integers
labels_dict= {'car1':0, 'car2':1, 'car3':2, 'car4':3, 'car5':4, 'car6':5}
```

```python
#to store the images as dataframe object
def dataframe(path):
    
    # Creating an empty list to store the image name and their labels.
    df = []
    
    
    for folder in os.listdir(path):
        # Define the path to the images
        imgs_path = path / folder
        
        # Get the list of all the images stored in that directory
        imgs = sorted(os.listdir(imgs_path))
        
        # Store each image path and corresponding label 
        for img_name in imgs:
            df.append((str(img_name), labels_dict[folder]))

        
    df = pd.DataFrame(df, columns=['image', 'Class'], index=None)
    
    # shuffle the dataset 
    df = df.sample(frac=1.).reset_index(drop=True)
    
    # Remove the extra columns and make data presentable
    i = df.columns.get_loc('image')
    df2 = df['image'].str.split("/", expand=True)
    df_1 = pd.concat([df.iloc[:, :i], df2, df.iloc[:, i+1:]], axis=1)
    #print(df_1)
    df = df_1.rename({0:'Image_ID'}, axis = 1)
    
            
    return df
```

```python
#train data set
# Path module of pathlib is used to store path
from pathlib import Path
data = Path(r'/content/sample_data/cars/train')

df = dataframe(data)
df.head()
```

```python
#Splitting this dataframe
df['split'] = np.random.randn(df.shape[0], 1)

msk = np.random.rand(len(df)) <= 0.7

train = df[msk]
train=train.drop(['split'], axis=1)
test = df[~msk]
test=test.drop(['split'], axis=1)
```

```python
print(train.head())
train.info()
```

```python
print(test.head())
test.info()
```

```python
#read the images based on Image_ID column
X = [ ]     # creating an empty array

for img_name in train.Image_ID:
    img = plt.imread('' + img_name)
    X.append(img)  # storing each image in array X
X = np.array(X)    # converting list to array
```

```python
#for test data
test_image = []
for img_name in test.Image_ID:
    img = plt.imread('' + img_name)
    test_image.append(img)
test_img = np.array(test_image)
```

```python
# one hot encoding Classes
from keras.utils import np_utils
train_y = np_utils.to_categorical(train.Class)
test_y = np_utils.to_categorical(test.Class)
```

```python
#Using a VGG16 pretrained model which takes an input image of shape (224 X 224 X 3) 
#Since our images are in a different size, we need to reshape
image = []
for i in range(0,X.shape[0]):
    a = resize(X[i], preserve_range=True, output_shape=(224,224)).astype(int)      # reshaping to 224*224*3
    image.append(a)
X = np.array(image)
```

```python
#for test
test_image = []
for i in range(0,test_img.shape[0]):
    a = resize(test_img[i], preserve_range=True, output_shape=(224,224)).astype(int)
    test_image.append(a)
test_image = np.array(test_image)
```

```python
# preprocessing the input data
from keras.applications.vgg16 import preprocess_input
X = preprocess_input(X,data_format=None)
test_image = preprocess_input(test_image, data_format=None)  
```

```python
#train test split to split images into train and validation sets
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, train_y, test_size=0.3, random_state=42)
```

```python
#building the model
#importing the libraries
from keras.models import Sequential
from keras.applications.vgg16 import VGG16
from keras.layers import Dense, InputLayer, Dropout
```

```python
#load the VGG16 pretrained model and store it as base_model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))    # include_top=False to remove the top layer
```

```python
#make predictions using this model for X_train and X_valid, get the features, and use them to retrain the model
X_train = base_model.predict(X_train)
X_valid = base_model.predict(X_valid)
test_image = base_model.predict(test_image)
X_train.shape, X_valid.shape, test_image.shape
```

```python
# reshaping -> converting to 1-D
X_train = X_train.reshape(102, 7*7*512)      
X_valid = X_valid.reshape(45, 7*7*512)
test_image = test_image.reshape(59, 7*7*512)
```

```python
# preprocessing
# centering the data - helps to converge model faster
train = X_train/X_train.max()      
X_valid = X_valid/X_train.max()
test_image = test_image/test_image.max()
```

```python
#build our model
# i. Building the model
model = Sequential()
model.add(InputLayer((7*7*512,)))    # input layer
model.add(Dense(units=1024, activation='sigmoid'))   # hidden layer
model.add(Dropout(0.5))      # adding dropout
model.add(Dense(units=512, activation='sigmoid'))    # hidden layer
model.add(Dropout(0.5))      # adding dropout
model.add(Dense(units=256, activation='sigmoid'))    # hidden layer
model.add(Dropout(0.5))      # adding dropout
model.add(Dense(6, activation='softmax'))            # output layer
```

```python
model.summary()
```

```python
# ii. Compiling the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# iii. Training the model
model.fit(train, y_train, epochs=100, validation_data=(X_valid, y_valid))
```
